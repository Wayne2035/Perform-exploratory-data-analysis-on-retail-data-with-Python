```python
import matplotlib.pyplot as plt
```

```python
Min_temp = [25,36,26,34,36,35]
Max_temp = [31,36,29,37,29,35]
Days = ["Sun","Mon", "Tue", "Wed", "Thurs", "Fri"]
plt.plot( Days, Min_temp, marker = "o", color = "red", linestyle = ":", markersize = 10, markerfacecolor = "blue", markeredgecolor = "green")
plt.plot( Days, Max_temp, marker = "*", color = "blue", linestyle = "-.", markersize = 10, markerfacecolor = "yellow", markeredgecolor = "orange")
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.legend(["Min_Tempt", "Max_Temp"])
plt.title("Daily Tempereture")
plt.grid(color = "pink", linestyle = "--", linewidth = 2, alpha = 0.5)
plt.show()
```

```python
courses = ["Python", "Java", "Data Science", "Kotlin", "Scala"]
Learners = [100,230,165,185,190]
colors = ["red", "blue", "orange", "cyan", "pink"]
plt.bar(courses, Learners, color = colors)
plt.xlabel("courses")
plt.ylabel("learners")
plt.grid(color = "pink", alpha = 0.4)
plt.title("Enrollment")
plt.show()
```

```python
Information = [10,12,14,16,18,19,17,21,11,55,66,32,25,44,11,52,85,25,68,25,25,3,65,2,66,35,45,55]
plt.hist(Information, color = "green")
plt.xlabel("Distribution")
plt.ylabel("Frequency")
plt.title("Histgram")
plt.grid(color = "pink", alpha = 0.3)
plt.show()
```

```python
## Scatter Plot
dataset1 = [10,20,25,30,35,45,65,75]
dataset2 = [15,18,20,25,28,30,32,42]
dataset3 = [20,22,24,30,36,46,50,55]
plt.scatter(dataset1, dataset2, color = "green",marker = "o", s = 80, edgecolor = "red")
plt.scatter(dataset1, dataset3, color = "red", marker = "s", s= 80, edgecolor = "yellow")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot")
plt.grid(color = "pink", alpha = 0.3, linestyle = ":")
plt.legend(["plot1", "plot2"])
plt.show()
```

```python
# Pie Chart
Company = ['Google', 'Microsoft', 'IBM', 'SalesForce', 'Meta']
Revenue = [100, 78, 34, 28, 77]

my_colors = ['red', 'cyan', 'magenta', 'blue', 'pink']
my_explode = [0.2,  0.2, 0.1, 0.1, 0.3]

plt.pie(Revenue, labels = Company, colors = my_colors, explode = my_explode, radius = 1.5
        , autopct = "%0.2f%%", shadow = True)
plt.show()
```
