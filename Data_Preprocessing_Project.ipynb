```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd. read_csv("/content/House Price_data cleaning project.csv")
```

```python
dataset.head()
```

```python
dataset.tail(10)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape[0]
```

```python
plt.figure(figsize = (30,30))
sns.heatmap(dataset.isnull())
plt.show()
```

```python
missing_percentage = (dataset.isnull().sum()/dataset.shape[0]) * 100
missing_percentage
```

```python
drop_columns = missing_percentage[missing_percentage >= 20].keys()
drop_columns
```

```python
dataset1 = dataset.drop(columns = drop_columns)
dataset1.shape
```

```python
dataset1.isnull().sum()
```

```python
dataset2 = dataset1.dropna()
```

```python
dataset2.isnull().sum()
```

```python
dataset2.shape
```

```python
num_columns = dataset2.select_dtypes (include = ["int64", "float64"]).columns
```

```python
num_columns
```

```python
print(len(num_columns))
```

```python
plt.figure(figsize = (30, 30))
for i, j in enumerate(num_columns):
    plt.subplot(19, 2,i+1)
    sns.histplot(dataset[j], bins=20, kde=True, color='blue', label='Dataset')
    sns.histplot(dataset2[j], bins=20, kde=True, color='orange', label='Dataset2')
    plt.title(j)
    plt.legend()
```
